```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.ensemble import RandomForestClassifier

```

```python
data = pd.read_csv("Dry_Eye_Dataset.csv")
data.head(5)
```

```python
data.info()
```

```python
# Select all columns with object (string) data type
# Replace "Y" with 1 and "N" with 0 in those columns

object_col=data.select_dtypes(include="object")
data[object_col.columns] = object_col.replace({"Y": 1, "N": 0})
```

```python
data
```

```python
data["Gender"]=data["Gender"].replace({"F":1,"M":0})
```

```python
data.head(5)
```

```python
##Converts categorical values into numbers
le=LabelEncoder()
data["Blood pressure"]=le.fit_transform(data["Blood pressure"])
```

```python
data.info()
```

```python
##Independent Variables
x=data.drop(["Dry Eye Disease"],axis=1)
##Dependent Variable(target for prediction)
y=data["Dry Eye Disease"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_test
```

```python
y_test
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
model=accuracy_score(y_test,y_pred)
```
